# 1. partition

```python
a = "abhishek.saini@gmail.com"

first = a.partition(".")[0].capitalize()
second = a.partition(".")[2].partition("@")[0].capitalize()

print(f"Hi {first} {second} , how are you")
print("Hello {} , {}".format(first,second))
```

# 2. split

```python
a = "abhishek.saini@gmail.com"

first = a.split(".")[0].capitalize()
second = a.split(".")[1].split("@")[0].capitalize()

print(f"Hi {first} {second} , how are you")
print("Hello {} , {}".format(first,second))
```
